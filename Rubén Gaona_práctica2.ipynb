```python
def dpar(M):
    dict_pares={}
    for par in M:
        if par%2==0:
            if par in dict_pares:
                dict_pares[par]+=1
            else:
                dict_pares[par]=1
    return dict_pares
```

```python
M = [12, 19, 19, 18, 18, 16, 18, 13, 19, 18, 12, 18, 11, 20, 14, 14, 20, 20, 20, 16, 13, 15, 19, 14, 12]
print(dpar(M))
```

```python
def TempMaxMin(lst):
    dict_ciudades={}
    for ciudades in lst:
        conteo=0
        for temperaturas in ciudades:
            if conteo>1:
                if temperaturas>max:
                    max=temperaturas
                if temperaturas<min:
                    min=temperaturas
            elif conteo==1:
                max=temperaturas
                min=temperaturas
            conteo+=1
        dict_ciudades[ciudades[0]]=[max, min]
    return dict_ciudades
```

```python
lst_ciudad = [['Londres', 3.4, 6.3, 10.5, 6.8], ['Oslo', -3.8, -5.0, 5.1, 4.2],
['Berlin', 7.5, 4.1, 12.3, 13.0], ['Málaga', 14.7, 12.3, 19.5, 18.4]]
TempMaxMin(lst_ciudad)
```

```python
def PersMayTemp(dPersC, dCiudT):
    lista_nombres=[]
    for clave_temperatura in dCiudT:
        if dCiudT[clave_temperatura]<0:
            for clave_persona in dPersC:
                if dPersC[clave_persona]==clave_temperatura:
                    lista_nombres.append(clave_persona)
    if not lista_nombres:
        return ""
    lista_ordenada=sorted(lista_nombres)
    return lista_ordenada
```

EJERCICIO 5

```python
def nivelKsang(dK, lst):
    dict_p={}
    for persona, K in dK.items():
        if K<lst[0]:
            dict_p[persona]="hipokalemia crítica"
        elif K<lst[1]:
            dict_p[persona]="hipokalemia leve"
        elif K<=lst[2]:
            dict_p[persona]="normal"
        elif K<=lst[3]:
            dict_p[persona]="hiperkalemia moderada"
        elif K>lst[3]:
            dict_p[persona]="hiperkalemia severa"
    return dict_p
```

```python
dK1 = {'Luis': 2.2, 'Carlos': 7.0, 'Laia': 4.0, 'Mikel': 5.5, 'Jordi': 5.2, 'Anna': 3.6, 'Joe': 7.2}
ls1 = [2.0, 3.5, 5.2, 7.0]
dic = nivelKsang(dK1, ls1)
print(dic)
```

```python
import pandas as pd
dhiper = {'Maria': [40, 135, 90],'Nuria': [63, 141, 92], 'Jose': [47, 110, 59], 'Luis': [49, 146, 94], 'Oriol': [52, 130, 89], 'Carlos': [65, 125, 89], 'Pepe': [70, 130, 92]}
listadedatos=[[nombre]+datos for nombre, datos in dhiper.items()]
Datos_arteriales= pd.DataFrame(listadedatos, columns=["Nombre", "Edad", "Sistólica", "Diastólica"])

def lst_hiper(dic):
    lista_nombres=[]
    for nombre, (edadp, sis, dias) in dic.items():
        if sis>=140 or dias>=90:
            estado="alta"
            lista_nombres.append(estado)
        elif sis<90 or dias<60:
            estado="baja"
            lista_nombres.append(estado)
        else:
            estado="normal"
            lista_nombres.append(estado)
    return lista_nombres

lista_nombres= lst_hiper(dhiper)

Datos_arteriales["Diagnóstico"]=lista_nombres
print(Datos_arteriales)
```
