```python
def FirstChar(s:str):
    if (s[0].isalpha()==True or s[0]=="_"):
        return True
    else:
        return False
```

```python
print(FirstChar('hola'))
print(FirstChar('_asdf13243'))
print(FirstChar('1ububu'))
```

EJERCICIO 2

```python
def porcentVocal(s:str):
    vocal=0
    if s=="":
        return 0.0
    for letra in s:
        if letra in "aeiouAEIOU":
            vocal+=1
    return round((vocal/len(s)*100), 1)
```

```python
print(porcentVocal('pepe'))
print(porcentVocal('ALEJandro'))
```

EJERCICIO 3

```python
def  nuevo_string(s:str, n:int):
    palabra=[]
    if n<0:
        return "El número debe ser mayor o igual a 0"
    for letra in s:
        if letra in "aeiouAEIOU":
            palabra.append(letra*n)
        else:
            palabra.append(letra)
    return "".join(palabra)
```

```python
nuevo_string('hola', 2)
nuevo_string('Charleston', 2)
```

EJERCICIO 4

```python
def  notas_al_pie(s:str):
    frase=[]
    pagina=0
    for letra in s:
        if letra == "*":
            pagina+=1    
            frase.append(f"({pagina})")
        else:
            frase.append(letra)
    return "".join(frase)
```

```python
notas_al_pie('Esta es la pagina * y esta la *')
```

EJERCICIO 5

```python
def codigo(s:str):
    if not s:
        return ""
    ini=[] #los apellidos
    count=0 #las letras de Nombres y Apellidos
    for letra in s:
        if letra.isupper() == True:
            ini.append(letra)
        if letra != " ":
            count+=1
    ini.append(f"{count}")
    return "".join(ini)
```

EJERCICIO 6

```python
def contar_hidrogenos(s:str):
    indice=0
    if s=="H":
            numhidrogenos=1
            return numhidrogenos
    for letra in s:
        if letra=="H":
            if indice+1==len(s):  #Si es una fórmula con un solo hidrógeno al final. Ejemplo: OH
                 numhidrogenos=1
                 return numhidrogenos
            elif s[indice+1].isalpha()==True:  #Si encuentra H y el siguiente valor es una letra hay 2 opciones: siguiente letra mayuscula, entonces es 1 Hidrogeno. Siguiente letra minuscula, entonces no es hidrogeno
                if s[indice+1].isupper()==True:
                    numhidrogenos=1
                    return numhidrogenos
            elif s[indice+1].isdecimal()==True: #Si encuentra H y hay un número después, ese es el número de Hidrógenos
                numhidrogenos=s[indice+1]
                return numhidrogenos
        indice+=1
```

```python
contar_hidrogenos("H20")
```

```python
contar_hidrogenos("H")
```

```python
contar_hidrogenos("Hg2H80")
```

```python
contar_hidrogenos("OH")
```

```python
def mediaTempRang(lst:list):
    temp1545=[]
    for temperatura in lst:
        if temperatura>=15.00 and temperatura<=45.00:
            temp1545.append(temperatura)
    if len(temp1545)==0:
        return -1
    return round(sum(temp1545)/len(temp1545),2)
```

```python
lst1 = [34.5, 12.9, 15, 43, 51.4, 23.4]
mediaTempRang(lst1)
```

```python
from math import log10
def SPL_dB(P):
    """ Calcula Sound Pressure Level en dB: 20log(x/20) x en microPascales """
    return 20*log10(P/20)
def detect2ndNdB(lst, N):
    aux=[]
    for n in lst:
        aux.append(SPL_dB(n))
    contador=0
    indice=0
    for decib in aux:
        if decib>=N:
            contador+=1
            if contador==2:
                return lst[indice]
        indice+=1
    return -1
```

```python
detect2ndNdB([90, 590, 750, 632, 650, 660, 2000, 789, 545], 30)
```

```python
detect2ndNdB([2000, 2450.5, 2500 , 456.7, 1567.8], 42)
```

```python
def es_primo(n):
    if n <= 1: return False
    for d in range(2, n//2+1):
        if n % d == 0:
            return False
    return True

def primoPitagoric2(lst):
    primos=[]
    primospitagoricos=[]
    contador=0
    for p in lst:
        if es_primo(p)==True:
            primos.append(p)
    for z in primos:
        if z%4==1:
            primospitagoricos.append(z)
            contador+=1
            if contador==2:
                return primospitagoricos
    if len(primospitagoricos)<2:
        return -1
```

```python
def contar_pos(m):
    contador=0
    for fila in m:
        for valor in fila:
            if valor>0:
                contador+=1
    return contador
```

```python
contar_pos([[1, -2, 3],[-4,5,6],[7,8,-9]])
```

```python
def mas_denso(Lst):                      #densidad=masa/volumen
    densidadmax=0
    for planeta in Lst:
        nombre, masa, volumen=planeta
        densidad=masa/volumen
        if densidad>densidadmax:
            densidadmax=densidad
            planetadenso=nombre
    return planetadenso
```

```python
mas_denso([['Marte', 1, 2], ['Tierra', 2, 3], ['Venus', 1, 3]])
```

```python
def jugComKm(equipo, x):
    jugadores=[]
    for i in range(0, len(equipo)):
        distancia=[]
        contador=0
        for datos in equipo[i]:
            if contador>=4 and equipo[i][2]==True:
                distancia.append(datos)
            contador+=1
        if len(distancia)!=0:
            media=sum(distancia)/len(distancia)
        else:
            media=0
        if media>x:
            jugadores.append(equipo[i][1])
    jugadoresordenados=sorted(jugadores)
    return jugadoresordenados
```
